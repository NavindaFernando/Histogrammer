```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('assets/Unequalized.jpg')
imgray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
print(img.shape)
print(imgray.shape)
```

### Image flatting

```python
flatten = imgray.flatten()
print(flatten.shape)
```

### Check unclear image histogram

```python
plt.hist(imgray.flatten(),bins=100)

plt.xlabel('pixel value')
plt.ylabel('occurence')
plt.show()
```

### Histogram equalization

```python
equ = cv2.equalizeHist(imgray)

cv2.imshow('img',imgray)
cv2.imshow('equ',equ)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Check equalized image histogram

```python
plt.hist(equ.flatten(),bins=100)
plt.xlabel('pixel value')
plt.ylabel('occurence')
plt.show()
```
